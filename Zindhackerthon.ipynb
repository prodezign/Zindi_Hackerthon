```python
cat_features = {
    "ward",
    "total_households",
    "total_individuals",
    "target",
}

for cat_feature in cat_features:
  print(cat_feature, train_df[cat_feature].unique(), sep=":")
  print("#"*50)
```

```python
train_df = train_df.drop(['ward', 'ADM4_PCODE'], axis=1)
test_df = test_df.drop(['ward', 'ADM4_PCODE'], axis=1)
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_y = train_df["target"] # Target
train_X = train_df.drop(("target"), axis=1)
```

```python
train_y.head()
```

```python
train_X.head()
```

```python
train_X_train, train_X_test, train_y_train, train_y_test = train_test_split(train_X, train_y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(train_X_train, train_y_train)
```

```python
reg.predict(train_X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from math import sqrt
y_pred_rmse_reg = reg.predict(train_X_test)

print("RMSE =",sqrt(mean_squared_error(train_y_test,y_pred_rmse_reg)))
```

```python
import xgboost as xgb
from xgboost import XGBRegressor

xgb_reg = XGBRegressor()
xgb_reg.fit(train_X_train, train_y_train)
y_pred_xgb = xgb_reg.predict(train_X_test)

print("RMSE =",sqrt(mean_squared_error(train_y_test,y_pred_xgb)))
```
